<h3>Artificial Neural Networks (ANNs)</h3>


An ANN has:
<ul>
    <li>An <em>input layer</em></li>
    <li>One of more <em>hidden layers</em></li>
    <li>An <em>output layer</em></li>
    </ul>
    

<h4>How a backprop ANN works</h4>
<ul>
    <li>Input presented to the model</li>
    <li>The values are calculated forward through the model to the output</li>
    <li>If we're learning, we estimate the <em>error</em> - the difference between the actual and calculated/predicted output.</li>
    <ul>
        <li>Activation Functions</li>
    </ul>
    <li>We propagate the error backwards through the model, and as part of this we change the weights to try and reduce the error from this particular input.</li>
    </ul>
<p> We repeat this process, first with each different piece of input data we have.
    <p>Then we show the whole data set over and over either until it performs to our satisfaction, or we decide it won't learn sufficiently, and we call it quits.
        

<h4>Recurrent Neural Networks</h4>
<p>A <em>recurrent neural network</em> (RNN) is one where the network knows what inputs it has seen before.
<p>(A normal ANN has no idea of anything but the current input.)
<p>In traditional RNNs:
    <ul>
        <li>The error signal tends to get weaker and weaker as subsequent inputs are seen.</li>
        <li>It is hard for an RNN to keep track of information/signal from  far away in a sequence of inputs (e.g. in a long paragraph, or a protein sequence).</li>
</ul>
<p> Newer models, <em>Transformers</em> have overcome many of these difficulties and form the basis for many of the state of the art AI/ML models (e.g. ChatGPT). 

<h4>Convolutional Neural Networks</h4>
<p>In earlier ANNs the architecture was relatively simple:
    <ul> 
        <li>All the hidden layers are the same, except for number of units</li>
        <li>All units in a layer were connected to all the units in the previous and following layer.</li>
        </ul>
<p>But this was considered to limit the abilities of the ANNs to learn more complex problems.
    <ul>
        <li>The model could easily <em>overfit</em> to the training data.</li>
        <li>As the models grow larger, this complete interconnection can result in large amounts of computation for all the weights between the layers.</li>
        </ul>
<p>In 1998, Yann LeCun published his first work on <em>Convolutional Neural Networks</em> (CNNs).
    <p>LeCun was inspired work on the neurons in the visual cortex.
<p>A CNN has different kinds of layers, and not all are completely connected:
    <ul>
        <li><em>Convolutional layer</em> - a layer that examines (/filters/scans) a part of the image (e.g., a 5x5 pixel segment).</li>
        <li>The convolutional layers learn the features of the image (e.g. the lines, edges, shapes) that make up the features relevant to learn the problem the model is trying to solve)</li>
        <li>Subsequent layers take these features and try and combine them to detect the presence or absence of higher level structures shapes, objects, etc.</li>
        <li>These are then <em>pooled</em> as input to the final hidden and output layers that are able to predict the problem the model is designed to solve.</li>
        </ul>

<h3>Deep Learning</h3>
<p>Deep Learning is based on supervised learning, specifically error backpropagation <em>artificial neural networks</em> (ANNs).
Normal Supervised Learning (e.g., ANNs):
    <ul>
        <li>Requires you to label data (e.g.,the images) you want to classify</li>
        <li>For all but simple problems, you need to do feature extraction on the images and provide those features as part of the imput.</li>
        </ul>
In contrast to Deep Learning:
   <ul>
    <li>It still needs the images to be labeled with the category you're tryiing to classify/predict</li>
    <ul>
        <li>But, wait - we may be able to get around that partially with <em>semi-supervised learning</em> which we'll discuss later.
    </ul>
    <li>We don't need to do feature extraction or add those extracted features to the input.</li>
    </ul>
    
<p>A Deep Learning system has several important differences from an ANN ML model.
    <ul>
        <li> It has more <em>layers</em>.</li>
        <li>You typically use a lot more data than with an ANN. </li>
        
        
    

<h4>Uses of Deep Learning</h4>
Ok, when do we use Deep Learning?
<p>Some considerations:
    <ul>
        <li>When it is worthwhile to collect the large amount of data a Deep Learning model needs</li>
        <li>When it is worthwhile to use the large amount of computation necessary to train a Deep Learning model.</li>
        <li>You might consider using cheaper, simpler ML models (e.g. ANNs) first, and then move to Deep Learing if those don't solve the problem.</li>
        <li>Sometimes even a Deep Learning model won't be able to solve a problem.</li>
        <li>For any kind of ML model, make sure your data is good (e.g., representative, covers all cases)</li>
        </ul>

<h4>Some things to think about</h4>
<ul>
    <li>Gary Marcus worries that Deep Learning does not get us to <em>Artificial General Intelligence.</em></li>
    <li>Deep Learning (as well as most ANNs and other ML models) tend to be <em>black boxes</em>/<em>opaque</em>. That is, it is hard to understand exactly how they solve their problems even when they work. To address this there is research going on in</li>
    <ul>
        <li>Explainable machine learning</li>
        <li>Trustworthy machine learning</li>
    </ul>
    <li>It takes <em>lots</em> of data to feed Deep Learning models.</li>
    <li>Some kinds of reasoning that deep learning does not (yet?) do well:</li>
        <ul>
    <li>hierarchical structure</li>
            <li>Open ended inference</li>
            <li>Conceptual thinking</li>
            <li>common sense</li>
            <li>Causation</li>
            <li>Prior Knowledge</li>
            <li>Static - I think this one may have been overcome (cf. work by DeepMind)</li>
            <li>The butterfly effect</li>
            <li>Overfitting</li>
    </ul>
</ul>

<h3>Generative Adversarial Networks (GANs)</h3>
<p>Invented by Ian Goodfellow
<p> Also one of the best technical textbooks on machine learning, <em>Deep Learning</em>
<p>A GAN has:
    <ul>
        <li>A <em>Generator</em> - creates (generates) possible examples of the problem of interest (e.g. lifelike cat pictures)</li>
        <li>A <em>Discriminator</em> - evaluates the generated examples to determine which are good and not.</li>
        <li>The model makes <em>Adjustments</em>The two parts of the GAN keep learning, hopefully eventually creating realistic images, etc. generation.</li>
            </ul>
<p>GAN-realted generative models underly all of the current technology that generates text (ChatGPY), images (DALL-E), etc.
<p>It also underlies <em>Deepfakes</em>
