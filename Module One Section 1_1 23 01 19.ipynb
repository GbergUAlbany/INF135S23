<b>C INF 135 Concepts of Artificial Intelligence (3 credits)</b>
<p>An introduction to the technical foundations of artificial intelligence as well as its relevance to, and implications for modern society. The course will cover the basics of artificial intelligence and machine learning (AI/ML). This will include hands-on work. The importance of data selection, handling, etc. will also be covered. In addition, as AI/ML becomes more prevalent, it has tremendous implications in the world. The course will introduce issues of fairness, bias, explainability, etc., and how these and other issues must be addressed in the design, deployment, and use of AI/ML systems.
<p><b>Prerequisite(s): None.</b>

<h2>Forward</h2>
"The adoption of Artficial Intelligence (AI) will be a major inflection point in human history." (p. vii)

It is revolutionary, because of the nature of the technology - like fire, the wheel, hydro/steam/internal combustion power, etc. But, also like those, "how it's administered and who has access to it will shape society for generations to come."

In the Forward, Vembu makes an interesting point I don't know that I agree with: 
<blockquote>"AI stands out from the other transformative technologies of the nineteenth and twentieth centuries, - the  steam engine, the electrical grid, genomics, computers, and the Internet - because it doesn't depend exclusively on critically expensive physical infrastructure to enable adotion; after all many of its benefits can be delivered through existing hardware we all carry about in our pockets."</blockquote>

Rather, he says the important point is that its effective adoption comes down to "intellectual infrastructure: education, understanding, and vision."

He also makes the point that it can cause tremendous political, social, and economic disruption.

In the longer term, Vembu views AI as the next step in a long trend towards automating repetitive tasks or those requiring tremendous strength.

But, the forward nicely encapsulates one of the themes of the course - the power of AI, and it's potential either for making our lives better or making them worse.


<h2>Introduction</h2>

Artifcial Intelligene in the Uber Example
<ul>
    <li> A Natural Language Processing (NLP) system </li>
    <li> Computer Vision Software </li>
    <li> Sensor data processing</li>
    <li> Algorithms to predict driver supply, rider demand, ETAs, etc.</li>
    </ul>

All this is necessary for an operation like Uber to operate at all. Without ths AI-based automoation, it could not operate at the scale it does.

This continues the digital automation trend we have seen since the middle ofthe 20th century. Computers have been doing more and more tasks that previously have required human calculation, etc.

The McDonalds example shows an interesting aspect of AI - how it can be used to <i>target</i> consumers. This has both positive and negative aspects.

"More and more jobs will require knowledge of AI. Granted, this does not mean you'll need ot learn programming languages or understand advanced statistics. But it will be critical to ha ea sold founation of the fundamentals." (p. x)
This course fits nicely in that space. 
<ul>
    <li>In it, we will try to learn about artificial intelligence (AI) and one of its big components, machine learning (ML). </li>
    <li>We will do a fair amount of hands on exercises. Some of which will involve programming. But, it is not at all a <i>programming course.</i></li>
    <li>We will look at both the technology of AI/ML, as well is its significance for all aspects of society.</li>
</ul>
Hopefully at the end of the course you will:
<ul>
    <li> Understand, at least schematically, the basics of AI/ML</li>
    <li> Be able to understand new advances in AI/ML</li>
    <li> Be able to make informed judgements about whether a statement or argument about AI/ML makes sense, or is wrong, flawed, hypred, or otherwise suspect.</li>
    <li> Be in a position to understand the importance of <i>data</i> for AI/ML work.</li>
    <li> Understand the ethical, societal, political,and economic implications of AI/ML.</li>
    <li> Be ready to take subsequent courses in the technology, societal aspects, ethics, <i>etc.</i> of AI/ML.</li>
</ul>
 
 I like the questions the author poses (p. xi):
 <ul>
    <li> Where does AI make sense?</li>
    <li> What are the gotchas?</li>
    <li> How do you evaluate the technology?</li>
    <li> What about starting an AI pilot?</li>
</ul>
These are just the kinds of judgements that I hope you can intelligently make after having taken this course.
